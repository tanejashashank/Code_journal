# To get a correlation matrix with heat map

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Shashank Taneja/Rotman/Fall term/Big Data Analytics Gerhard RSM 8413/Assignments/Group assignment due 20-10-19/UniversalBank_Data.csv")
```

```python
corrs = data.corr()
mask = np.zeros_like(corrs)
mask[np.triu_indices_from(mask)] = True
sns_plot = sns.heatmap(corrs, cmap='Spectral_r', mask=mask, square=True, vmin=-.4, vmax=.4)
plt.title('Correlation matrix')
fig = sns_plot.get_figure()
fig.savefig("output.png")
```
